```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
       L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if(cnt < 3):
           L.append(i['name'])
           cnt += 1
        else:
           break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.iloc[0].cast
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
       if(i['job'] == 'Director'):
          L.append(i['name'])
          break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
new_cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
new_vectors = new_cv.fit_transform(new_df['tags']).toarray() #list to array
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(new_vectors)
```

```python
similarity[0]
```

```python
new_df[new_df['title'] == 'Spectre'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)  
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
